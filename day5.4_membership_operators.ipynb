### Membership operators
- **Membership operators** in Python check whether a value **exists inside a group of items** like a list, string, or tuple. The two operators are **`in`** and **`not in`**. They return **True or False** depending on whether the item is present.

**Example:**

```python
fruits = ["apple", "banana", "mango"]
print("apple" in fruits)
```
