## Assignment operators


```python
#Exercise 1: Basic Assignment
#Task
#Assign the value 20 to a variable and print it.
a = 10
print(a)
```

```python
#Exercise 2: Add and Assign (+=)
#Task
#Start with 10, add 5 using +=.
a = 10
a += 5
print("updated value:",a)
```

```python
#Exercise 3: Subtract and Assign (-=)
#Task
#Start with 50, subtract 20 using -=.
a = 50
a -= 20
print("updated value of a :",a)
```

```python
#Exercise 4: Multiply and Assign (*=)
#Task
#Start with 8, multiply by 3.
a = 8
a *=3
print("updated value is:",a)
```

```python
#Exercise 5: Divide and Assign (/=)
#Task
#Start with 40, divide by 6
a = 40
a /= 6
print("updated value:",a)
```

```python
#Exercise 6: Floor Divide (//=)
#Task
#Start with 17, floor divide by 3.
a = 17
a //=3
print("updated value:",a)
```

```python
#Exercise 7: Modulus Assign (%=)
#Task
#Start with 20, find remainder when divided by 6.
a = 20
a %= 6
print("updated value:",a)
```

```python
#Exercise 8: Power Assign (**=)
#Task
#Start with 4, find square using **=.
a = 10
a **= 2
print("updated value:",a)
```
