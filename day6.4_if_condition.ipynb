### if condition
An **if condition** in Python is used to **make decisions**. The program checks a condition, and if it is **True**, it runs certain code.

**Example:**

```python
age = 20
if age >= 18:
    print("You can vote")
```

Here, Python checks the age. If it is 18 or more, it prints the message.
