```python
#Exercise 5: Simple Interest Calculator
#Task:
#Take Principal, Rate, Time from user
#Calculate simple interest
#Formula = (P * R * T) / 100
principal = float(input("enter principal amount"))
rate = float(input("enter rate of interest"))
time = float(input("enter time in years"))
simple_interest = (principal*rate*time)/100
print("simple interest is:",simple_interest)
```

```python
#Exercise 6: Boolean Practice
#Task:
#Ask: "Are you a student? (yes/no)"
#Convert it into True or False
#Print the boolean value
say_answer = input("are you a student? (yes/no)")
if say_answer.lower() == "yes":
    is_student = True
else:
    is_student = False
print("Is the person a student",is_student)
print(type(is_student))
```
