### Binary operators
- **Bitwise operators** in Python work on numbers at the **binary level (0s and 1s)** instead of normal decimal numbers.

**Types:**

* `&` AND
* `|` OR
* `^` XOR
* `~` NOT
* `<<` Left shift
* `>>` Right shift

**Example:**
`5 & 3` → result is **1** because in binary `101 & 011 = 001`.


```python
#Exercise 1: Bitwise AND (&)
#Task
#Find the bitwise AND of two numbers.
#Example numbers
#5 and 3
a = 5
b = 3
print(a&b)
```

```python
#Exercise 2: Bitwise OR (|)
#Task
#Find bitwise OR of two numbers.
a = 5
b = 3
print(a|b)
```

```python
#Exercise 3: Bitwise XOR (^)
#Task
#Find bitwise XOR of two numbers.
a = 5
b = 3
print(a^b)
```

```python
#Bitwise NOT (~)
#Task
#Find bitwise NOT of a number.
num = 5
print("The not of a number is:",~num)
```

```python
#Exercise 5: Left Shift (<<)
#Task
#Left shift number 3 by 2 positions
a = 3
print("a left shift by 2 positions:",a<<2)
```

```python
#Exercise 6: Right Shift (>>)
#Task
#Right shift 16 by 2 positions
a = 16
print("a right shift by 2 positions:",a>>2)
```
