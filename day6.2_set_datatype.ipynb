### set (data type)
- A **set** in Python is a collection of **unique items**, meaning it automatically removes duplicates. It uses **curly braces `{}`**. Sets are useful when you only want **distinct values**.

**Example:**

```python
numbers = {1, 2, 2, 3}
print(numbers)
```

Output: `{1, 2, 3}` because duplicate `2` is removed automatically.

````python
#Exercise 3: Add an Element
#Task
#Add "orange" to a set of fruits.
fruits = {"apple","banana","grapes","orange"}
fruits.add("guava")
fruits.add("strawberry")
fruits.add("apple")
print(fruits)
````

````python
#Exercise 4: Remove an Element
#Task
#Remove "banana" from a set.
fruits = {"apple","banana","grapes","orange"}
fruits.remove("banana")
print(fruits)
````

````python
#Exercise 5: Check if Element Exists
#Task
#Check if "apple" exists in the set.
fruits = {"apple","banana","grapes","orange"}
print("apple" in fruits)
````

````python
#Find Length of Set
#Task
#Find the number of elements in a set.
my_numbers = {1,2,3,4,5,5,4,3,2,1}
print("The length of list is:",len(my_numbers))
````

````python
#Exercise 7: Union of Two Sets
#Task
#Combine two sets.
set_1 = {1,2,3,4}
set_2 = {5,6,7,3,8}
print("The union of 2 sets is :",set_1.union(set_2))
````

````python
#Exercise 8: Intersection of Two Sets
#Task
#Find common elements between two sets.
set_1 = {1,2,3,4,5}
set_2 = {3,4,5,6,7}
print("The intersection of 2 numbers is:",set_1.intersection(set_2))
````

````python
#Exercise 9: Difference Between Sets
#Task
#Find elements present in set1 but not in set2.
set_1 = {1,2,3,4,5}
set_2 = {3,4,5,6,7}
print("the difference of 2 sets is:",set_1.difference(set_2))
````

````python
#Exercise 10: Convert List to Set
#Task
#Convert a list with duplicates into a set.
my_list = [1,2,3,4,4,5,2,3,5]
my_set = set(my_list)
print(my_set)
print(type(my_set))
````
