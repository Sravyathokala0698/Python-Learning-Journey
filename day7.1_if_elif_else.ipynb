### if elif else
- An **if-elif-else** condition lets a program choose between **multiple possibilities**. Python checks conditions one by one. If the first `if` is false, it checks `elif`. If none are true, `else` runs.

**Example:**

```python
marks = 75
if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```


````python
#Exercise 1: Grade Calculator
#Task
#Take marks from the user and print the grade.
#Marks ≥ 90 → Grade A
#Marks ≥ 60 → Grade B
#Otherwise → Grade C
marks = int(input("enter the marks:"))
if marks>=90:
    print("Grade A")
elif marks>=60:
    print("Grade B")
else:
    print("Grade C")
````
