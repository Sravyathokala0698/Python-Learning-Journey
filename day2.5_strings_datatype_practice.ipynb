## Strings

```python
##Exercise 1: Create and Print a String
#Task:
#Create a variable name
#Print it
#Print its type
name = "sravya"
print(name)
print("type of name is:",type(name))
```

```python
##Exercise 2: String Length
#Task:
#Take a name from user
#Print number of characters
course = "data science and machine learning"
print(len(course))
```

```python
#Exercise 3: String Indexing
#Task:
#Create a string "Python"
#Print:
#First character
#Last character
name = "python"
print(name)
#first character
print(name[0])
#last character
print(name[-1])
```

```python
#Exercise 4: String Slicing
#Task:
#From "DataScience", 
#print:
#"Data"
#"Science"
course_name = "Datascience"
print(course_name[0:4])
print(course_name[4:11])
```

```python
#Exercise 5: Convert Case
#Task:
#Take input from user
#Print:
#Uppercase
#Lowercase
input_user = "SravyaThokala"
print(input_user.lower())
print(input_user.upper())
```

```python
#Exercise 6: Remove Spaces
#Task:
#Remove extra spaces from beginning and end and not in between
course_name= " Data science and manchine learning   "
print(course_name.strip())
```

```python
#Exercise 7: Replace Word
#Task:
#Replace "bad" with "good"
course1 = "The couse is bad"
course2 = course1.replace("bad","good")
print(course2)
```

```python
#Exercise 8: Check Substring
#Task:
#Check if "Data" is present in "Data Science"
#this is membership operator
text = "Data Science"
print("Data" in text)
```

```python
#Exercise 9: Count Characters
#Task:
#Count how many times "a" appears in "banana"
word = "banana"
print(word.count("a"))
```

```python
#Exercise 10: Split String
#Task:
#Split "apple,banana,mango" into list
fruits ="apple,banana,mango"
fruits_list =fruits.split(",")
print(fruits_list)
print(type(fruits_list))
```
