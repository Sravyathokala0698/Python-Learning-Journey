### list data types

```python
#Exercise 1: Create and Print a List
#Task:
#Create a list of 5 numbers
#Print the list
#Print its type
numbers = [1,2,3,4,5]
print(numbers)
print(type(numbers))
```

```python
#Exercise 2: Access Elements (Indexing)
#Task:
#From list [100, 200, 300, 400] print:
#First element
#Last element
numbers = [100,200,300,400]
print(numbers[0])
print(numbers[-1])
```

```python
#Exercise 3: Modify List Element
#Task:
#Change second element of [1, 2, 3, 4] to 20.
number_1 = [1,2,3,4,5]
number_1[1]=20
print(number_1)
```

```python
#Exercise 4: Add Elements
#Task:
#Add 50 to the end of list [10, 20, 30]
#Add 5 at beginning
number_2 = [10,20,30,40]
number_2.append(50)
number_2.insert(0,5)
print(number_2)
```

```python
#Exercise 5: Remove Elements
#Task:
#From [10, 20, 30, 40]
#Remove 30
#Remove last element
numbers_3 =[10,20,30,40,50]
numbers_3.remove(30)
print(numbers_3)
numbers_3.pop()
#print(numbers_3)
```

```python
#Exercise 6: Length of List
#Task:
#Find number of elements in list.
fruits = ["apple","banana","grapes","orange"]
print("length of list is:",len(fruits))
```

```python
#Exercise 7: Sum of List
#Task:
#Find sum of numbers in list.
numbers_4 = [1,2,3,4,5]
print("sum of the numbers is:",sum(numbers_4))
```

```python
#Exercise 8: Take List Input from User
#Task:
#Ask user to enter numbers separated by comma
#Example: 10,20,30
#Convert into list
#Print the list
numbers = input("enter numbers seperated by comma:")
num_list = numbers.split(",")
print(num_list)
```

```python
#Exercise 9: Check if Item Exists
#Task:
#Check if "apple" exists in list.
fruits = ["apple","mango","grapes","orange"]
print("apple" in fruits)
```

```python
#Exercise 9: Convert String List to Integer List
#Task:
#Convert list elements to integers.
#print sum of integer list
numbers = input("enter numbers seperated by comma")
num_list=numbers.split(",")
print(num_list)
#converting string list to integer list
int_list = [int(num) for num in  num_list]
print("enter integer list:",int_list)
print("sum of integers is:",sum(int_list))
```
