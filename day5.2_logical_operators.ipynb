## logical operators

```python
#(and , or, not) are called logical operators
#Exercise 1: AND Operator
#Task
#Take age and check if a person is eligible to vote and eligible for senior benefits.
#Condition:
#Age ≥ 18
#Age ≤ 60
age = int(input("enter the age:"))
print("Is the person eligible for voting:",(age>=18 and age<=60))
```

```python
#Exercise 2: OR Operator
#Task
#Take marks and check if the student passed in at least one subject.
#Condition:
#Math ≥ 40 OR Science ≥ 40
maths = int(input("enter the maths marks:"))
science = int(input("enter the science marks:"))
print("Is the student passed in atleast one exam:",(maths>=40 or science>=40))
```

```python
#Exercise 3: NOT Operator
#Task
#Check if a number is not greater than 50.
num_1 = int(input("enter the name 1:"))
print("The number is not greater than 50:",not(num_1>50))
```

```python
#Exercise 4: Age Range Check
#Task
#Check if a person’s age is between 20 and 30.
age = int(input("enter the age:"))
print("is the person age between 20 and 30:",(age>=20 and age<=30))
```

```python
#Exercise 5: Salary Eligibility
#Task
#Check if salary is either very low or very high.
#Condition:
#Salary < 20000 OR Salary > 100000
salary = int(input("enter the salary:"))
print("Is the salary too low or too high:",(salary<20000 or salary>100000))
```

```python
#Exercise 6: Login Check
#Task
#Check login eligibility.
#Condition:
#Username = "admin"
#Password = "1234"
#Both must be correct.
user_name = input("enter user name:")
pass_word = int(input("enter the password:"))
print("The atm user name and password are correct:",(user_name=="admin" and pass_word==1234))
```

```python
#Exercise 7: Number Range Check
#Task
#Check if a number is not between 10 and 20.
num_1 = int(input("enter a number:"))
print("The number is not between 10 and 20:",not(num_1>=10 and num_1<=20))
```
