### for loop
- A **for loop** in Python is used to **repeat an action for each item in a sequence**, like a list, string, or range. It automatically goes through each value one by one.

**Example:**

```python
for i in range(3):
    print(i)
```

This prints **0, 1, 2**, repeating the print statement three times.


````python
#Exercise 1: Print Numbers 1–5
#Task
#Print numbers from 1 to 5 using a for loop.
for i in range(1,6):
    print(i)
````

````python
#Exercise 7: Count Numbers
#Task
#Count how many numbers are in a list.
numbers =[10,20,30,40]
count = 0
for num in numbers:
    count += 1
print("total numbers count is:",count)
````

````python
#Exercise 8: Find Largest Number
#Task
#Find the largest number in a list.
numbers =[10,20,30,40,70,88,100,240]
largest = numbers[0]
for num in numbers:
    num>largest
    largest=num
print("The largest number in the list is :",largest)
````
