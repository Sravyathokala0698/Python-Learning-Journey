### Nestef If
- A **nested if condition** means placing an **`if` statement inside another `if` statement**. It is used when a second condition should be checked **only if the first condition is true**.

**Example:**

```python
age = 20
if age >= 18:
    if age >= 21:
        print("Eligible for everything")
```


````python
#Exercise 1: Voting + Driving Eligibility
#Task
#Check if a person:
#Age ≥ 18 → Can vote
#Age ≥ 21 → Can also drive
age = int(input("enter the age:"))
if age >=18:
    print("eligible to vote")
    if age>=21:
         print("eligible for driving")
    else: 
         print("not eligible for driving")
else:
    print("not eligibe to vote")
````

````python
#Exercise 2: Pass and Distinction
#Task
#Check marks:
#Marks ≥ 40 → Pass
#Marks ≥ 75 → Distinction
marks = int(input("enter the marks"))
if marks>=40:
    print("pass")
    if marks>=75:
        print("Distinction")
    else: 
        print("not distinction")
else:
    print("fail")
````
