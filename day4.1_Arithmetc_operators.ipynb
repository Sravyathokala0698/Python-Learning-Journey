```python
#Exercise 1: Basic Addition
#Task:
#Take two numbers from user and print their sum.
num_1 = int(input("enter number 1:"))
num_2 = int(input("enter number 2:"))
sum = num_1+num_2
print("The sum of 2 numbers is:",sum)
```

```python
#Exercise 2: Subtraction & Multiplication
#Task:
#Take two numbers and print:
#Difference
#Product
num_1 = int(input("enter number 1:"))
num_2 = int(input("enter number 2:"))
diff = num_1-num_2
print("The difference of the numbers is:",diff)
prod = num_1*num_2
print("The product of the numbers is:",prod)
```

```python
#Exercise 3: Division vs Floor Division
#Task:
#Take two numbers and print:
#Normal division (/)
#Floor division (//)
num_1 = int(input("enter number 1:"))
num_2 = int(input("enter number 2:"))
normal_div = num_1/num_2
print("The result of normal division is:",normal_div)
floor_div = num_1//num_2
print("The result of floor division is:",floor_div)
```

```python
#Exercise 4: Modulus (Remainder)
#Task:
#Check remainder when a number is divided by 2.
num_1 = int(input("enter number 1:"))
rem = num_1%2
print("The remainder when divided by 2 is:",rem)
```

```python
#Exercise 5: Power Operator
#Task:
#Take a number and print:
#Square
#Cube
num_1 = int(input("enter number 1:"))
square_2 = num_1**2
cube_3 = num_1**3
print("The square of a number is:",square_2)
print("The cube of a number is:",cube_3)
```

```python
#Exercise 6: Simple Interest Calculator
#SI = (P * R * T) / 100
p = float(input("enter principal amount:"))
t = float(input("enter time:"))
r = float(input("enter rate of interest:"))
si = (p*t*r)/100
print("the simple interest is:",si)
```

```python
#Exercise 7: Area of Circle
#Area = π * r²
radius_1 =float(input("enter the radius of circle:"))
area_1 = 3.14*(radius_1)**2
print("The area of a circle is:",area_1)
```

```python
#Exercise 8: Temperature Converter
#Celsius → Fahrenheit
#(C × 9/5) + 32
c = float(input("enter the celsius:"))
fahrenheit_1 = (c * 9/5) + 32
print("The celsius in fahrenheit is:",fahrenheit_1)
```

```python
#Exercise 9: Salary Calculation
#Task:
#Take salary and bonus percentage.
#Calculate final salary.
salary = int(input("enter the salary:"))
bonus_percentage = int(input("enter bonus percentage:"))
bonus_amount = (salary/bonus_percentage)
print("The bonus amount is:",bonus_amount)
total_salary = salary+bonus_amount
print("The total salary is:",total_salary)
```
