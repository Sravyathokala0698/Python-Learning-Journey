### if-else condition
- An **if-else condition** in Python helps a program **choose between two actions**. If the condition is **True**, one block of code runs. If it is **False**, another block runs.

**Example:**

```python
age = 16
if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")
```

Here Python decides based on age.
