### Identity Operators
- **Identity operators** in Python check whether **two variables refer to the same object in memory**, not just whether their values are equal. The operators are **`is`** and **`is not`**.

**Example:**

```python
a = [1, 2]
b = a

print(a is b)   # True
```
Here `a` and `b` point to the **same list object**, so `a is b` is **True**.


````python
#Exercise 1: Basic Identity Check
#Task
#Create two variables with the same value and check if they are the same object.
a = 10
b = 20
print(a is b)
````

````python
#Exercise 2: Compare Two Lists
#Task
#Create two lists with the same values and check:
#Are they equal?
#Are they the same object?
list_1 = [1,2,3]
list_2 = [1,2,3]
print("equal values:",list_1 == list_2)
print("same object:",list_1 is list_2)
````

````python
#Exercise 3: Same Object Reference
#Task
#Make two variables refer to the same list.
list_1 = [1,2,3]
list_2 = list_1
print("same-object:",list_1 is list_2)
````

````python
#Exercise 4: Using is not
#Task
#Check if two variables are not the same object.
list_1 = [1,2,3]
list_2 = list_1
print("using is not:",list_1 is not list_2)
````

````python
#Exercise 5: None Check (Very Common in Python)
#Task
#Check if a variable is None.
value = None
print(value is None)
````

````python
#Exercise 6: Identity with Strings
#Task
#Check identity of two strings.
string_1 = "python"
string_2 = "python"
#print(string_1==string_2)
print(string_1 is string_2)
````

````python
#Exercise 7: Identity with User Input
#Task
#Take two numbers from the user and check if they refer to the same object.
a = int(input("enter number 1"))
b = int(input("enter number 2"))
print("same object:",a is b)
````
