# Dictionary Datatypes
- A **dictionary** in Python stores data as **key–value pairs**, like a real dictionary where a word has a meaning. Each key is unique and used to access its value. It uses **curly braces `{}`**.

**Example:**

```python
person = {"name": "Shravya", "age": 25}
print(person["name"])
```

Output: **Shravya**.


````python
#Exercise 1: Create and Print a Dictionary
#Task
#Create a dictionary that stores name, age, and city and print it.
person ={"name":"sravya","age": 27, "course":"data science"}
print(person)
print(type(person))
````

````python
#Exercise 2: Access Dictionary Value
#Task
#Print the age from the dictionary.
person_details ={"name":"sravya","age":27}
print(person_details["age"])
````

````python
#Exercise 3: Add a New Key-Value Pair
#Task
#Add "profession": "Data Scientist" to the dictionary.
person ={"name":"sravya","age":27}
person["course"]="data science"
print(person)
````

````python
#Exercise 4: Update a Value
#Task
#Change age from 25 → 26.
person ={"name":"sravya","age":27}
person["age"]=25
print(person)
````

````python
#Exercise 5: Remove an Item
#Task
#Remove "city" from the dictionary.
person ={"name":"sravya","age":27,"city":"hyderabad"}
person.pop("city")
print(person)
````

````python
#Exercise 6: Print Keys and Values
#Task
#Print all keys and all values.
person ={"name":"sravya","age":27}
print("keys are:",person.keys())
print("values are:",person.values())
````

````python
#Exercise 7: Check if Key Exists
#Task
#Check if "name" exists in the dictionary.
person ={"name":"sravya","age":27}
print("name" in person)
````

````python
#Exercise 8: Loop Through Dictionary
#Task
#Print each key and value.
person ={"name":"sravya","age":27,"course":"data science"}
for key,value in person.items():
    print(key,":",value)
````

````python
#Exercise 9: Dictionary Length
#Task
#Find how many key-value pairs exist.
person ={"name":"sravya","age":27,"course":"data science"}
print(len(person))
````

````python
#Exercise 10: Create Dictionary from User Input
#Task
#Take name and age from user and store in dictionary.
name = input("enter your name:")
age = int(input("enter your age:"))
course =input("enter course name:")
person ={"name":name,"age":age,"course":course}
print(person)
````
