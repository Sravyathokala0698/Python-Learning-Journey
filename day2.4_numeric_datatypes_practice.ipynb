```python
# In Python, numeric data types are:
#int (Integer)
#float (Decimal numbers)
#complex (Complex numbers)
```

```python
#Exercise 1: Integer Practice
#Task:
#Create two integer variables
#Add, subtract, multiply them
#Print results and types
num_1 = int(input("enter value of number 1"))
num_2 = int(input("enter value of number 2"))
print("The addition is:",(num_1+num_2))
print("The subtraction is:",(num_1-num_2))
print("The multiplication is:",(num_1*num_2))
```

```python
#Exercise 2: Float Practice
#Task:
#Create two float numbers
#Divide them
#Print result and type
num_1 = float(input("enter a number"))
num_2 = float(input("enter another number"))
div = num_1/num_2
print("result is",div,type(div))
```

```python
#Exercise 3: Integer vs Float Division
#Task:
#Check difference between / and //
num_1 = 50
num_2 =23
diff_1 = num_1/num_2
print(diff_1,type(diff_1))
floor_div =num_1//num_2
print(floor_div,type(floor_div))
```

```python
#Exercise 4: Power and Modulus
#Task:
#Find square of a number
#Find remainder when divided by 3
num_1 = 5
num_2 = (num_1)**2
print("The square of a number is:",num_2)
#find remainder when divided by 3
print("The remainder is",(num_1%5))
```

```python
#Exercise 5: Complex Numbers
#Task:
#Create a complex number
#Print real and imaginary parts
a = 4+3j
print("The real part is:",a.real)
print("The imaginery part is:",a.imag)
print(type(a))
```

```python
#Exercise 6: Convert Between Numeric Types
#Task:
#Convert int → float
#Convert float → int
#Convert int → complex
a = 20
b = 15.3
c = 50
#converting int to float
d = float(a)
print("type of d is:",d,type(d))
#converting float to int
e = int(b)
print("type of e is",e,type(e))
#converting int to complex
f = complex(c)
print("type of f is:",f,type(f))

```

```python
#Exercise 7: User Input Numeric Practice
#Task:
#Take two numbers from user
#Convert to float
#Print:
#Sum
#Difference
#Product
#Division
num_1 = int(input("enter any number1:"))
num_2 = int(input("enter any number2:"))
num_3 = float(num_1)
num_4 = float(num_2)
print(num_3)
print(num_4)
sum = num_3+num_4
diff = num_3-num_4
mul = num_3*num_4
div =num_3/num_4

print("The addition of two numbers is:",sum)
print("The subtraction of two numbers is:",diff)
print("The multiplication of two numbers is:",mul)
print("The division of two numbers is:",div)
```
