```python
#TASK: Student Marks Analyzer
#Problem Statement
#Write a Python program that:
#Takes marks as input separated by comma
#Example:45,67,89,23,90,55
#Convert them into integer list
#Print:
#1.Total marks
#2.Average marks
#3.Highest mark
#4.Lowest mark
#5.Number of students who scored above 60
#6.Create a new list containing only marks above average
marks = input("Take marks input from the user seperated by comma:")
#(student1,student2,student3,student4,student5,student6)
str_marks = marks.split(",")
print(str_marks)
#converting string to int
int_marks = [int(mark) for mark in str_marks]
print(int_marks)
#1.total marks -> print sum
sum_total=sum(int_marks)
print("The total marks is:",sum_total)
#2. average of marks
avg_marks = sum_total/len(str_marks)
print("The avg of marks is:",avg_marks)
#3. highest marks 
high_marks=max(int_marks)
print("The highesh marks is:",high_marks)
#4. minimum marks
min_marks= min(int_marks)
print("The minimum of the marks is:",min_marks)
#5.number of students who scored above 60
```
