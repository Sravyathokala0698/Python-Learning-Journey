### Comparison Operators

```python
#Exercise 1: Greater Than
#Task
#Take a number from the user and check if it is greater than 50.
num_1 = int(input("enter the number:"))
print("The is greater than 50",num_1>50)
```

```python
#Exercise 2: Compare Two Numbers
#Task
#Take two numbers and check if the first number is greater than the second.
num_1 = int(input("enter number 1:"))
num_2 = int(input("enter number 2:"))
print("number 1 is greater than number 2:",num_1>num_2)
```

```python
#Exercise 3: Equality Check
#Task
#Take two numbers and check if they are equal.
num_1 = int(input("enter number 1:"))
num_2 = int(input("enter number 2:"))
print("Both the numbers are equal:",num_1==num_2)
```

```python
#Exercise 4: Not Equal Operator
#Task
#Take two numbers and check if they are not equal.
num_1 = int(input("enter number 1:"))
num_2 = int(input("enter number 2:"))
print("Both the numbers are not equal:",num_1!=num_2)
```

```python
#Exercise 5: Age Eligibility
#Task
#Take age from user and check if the person is eligible to vote (age ≥ 18).
num_1 = int(input("enter the age of the person:"))
print("Is the person eligible to vote:",num_1>18)
```

```python
#Exercise 6: Pass or Fail
#Task
#Take marks from user and check if the student passed (marks ≥ 40).
marks = int(input("enter marks of the user:"))
print("The student is passed:",marks>=40)
```

```python
#Exercise 7: Range Check
#Task
#Check if a number is less than or equal to 100.
num_1 = int(input("enter number 1:"))
print("If the number is less than or equal to 100:",num_1<=100)
```

```python
#Exercise 8: Compare String Length
#Task
#Take two names and check which name is longer.
name_1 = input("enter the name 1")
name_2 = input("enter the name 2")
print("name 1 is longer than name 2:",len(name_1)>len(name_2))
```
