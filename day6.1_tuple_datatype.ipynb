###  Tuple
- A **tuple** in Python is a collection of values stored together like a list, but **it cannot be changed after creation** (immutable). It uses **parentheses `()`**.

**Use case:** storing fixed data.

**Example:**

```python
person = ("Shravya", 25, "Data Science")
```

Here the name, age, and field stay constant and shouldn’t be modified.


````python
#Exercise 1: Create and Print a Tuple
#Task
#Create a tuple of 4 numbers and print it.
numbers = (1,2,3,4)
print(numbers)
print(type(numbers))
````

````python
#Exercise 2: Access Tuple Elements
#Task
#Print the first and last element of a tuple.
num = (1,2,3,4,5)
fruits = ("apple" , "mango" ,"grapes")
print(num[0])
print(num[-1])
print(fruits[0])
print(fruits[-1])
````

````python
#Exercise 3: Tuple Length
#Task
#Find how many elements are in a tuple.
my_tuple = (1,2,3,4,5)
print("The length of tuple is:",len(my_tuple))
````

````python
#Exercise 4: Check if Item Exists
#Task
#Check if "banana" exists in a tuple
fruits = ("apple", "banana" ,"grapes")
print("apple" in fruits)
````

````python
#Exercise 5: Tuple Slicing
#Task
#Print the first 3 elements from a tuple.
my_tup = (1,2,3,4,5,6,7,8,9)
print(my_tup[0:3])
````

````python
#Exercise 6: Count Elements
#Task
#Count how many times 2 appears in the tuple.
my_tuple = (1,2,2,3,2,4,2,5,2,6,7,2,2,2,9)
print(my_tuple.count(2))
````

````python
#Exercise 7: Find Index
#Task
#Find the index of "mango".
fruits = ("apple" ,"mango" ,"grapes" ,"orange")
print(fruits.index("mango"))
````

````python
#Exercise 8: Convert List to Tuple
#Task
#Convert a list into a tuple.
my_list = [1,2,3,4,5]
print(my_list)
my_tuple = tuple(my_list)
print(my_tuple)
print(type(my_tuple))
````

````python
#Exercise 9: Tuple Unpacking
#Task
#Assign tuple values to variables.
my_tuple = ("sravya", 27 ,"data science")
name,age,course = my_tuple
print(name)
print(age)
print(course)
````

````python
#Exercise 10: Loop Through Tuple
#Task
#Print each element using a loop.
my_tuple = (1,2,3,4,5)
for num in my_tuple:
    print(num)
````
